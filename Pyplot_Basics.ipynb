```
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```
x = ['A','B','C','D','E','F']
y1 = [25,12,30,18,6,23]
y2 = [16,28,23,11,9,14]
```

```
plt.plot(x,y1)
plt.show()
```

```
# Cada linha recebe uma cor automaticamente caso não especificado
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```



---


**Para configurar uma série de dados, adicionamos um terceiro parâmetro com as informações de:**
* Tipo de Marcador
* Tipo de Linha
* Cor

[Opções de configuração](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)

```
plt.plot(x,y1,'--r') #sem marcador, tracejado e vermelho
plt.plot(x,y2,'*:g') #marcador estrela, pontilhado e verde
plt.show()
```
