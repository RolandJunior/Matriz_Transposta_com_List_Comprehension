```
import pandas as pd
import matplotlib.pyplot as plt
```

```
plt.rcParams["figure.figsize"] = (14,6)
```

```
arquivo = '../content/drive/My Drive/Temp/salarios2.csv'
df = pd.read_csv(arquivo)
df.head()
```

```
subset = df[df['Department']=='POLICE']['Annual Salary']
subset.hist(bins=40)
plt.show()
```

```
subset = df.groupby('Department')['Annual Salary'].mean()
subset.plot(kind='bar')
plt.show()
```
