```
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```
x = ['A','B','C','D','E','F']
y1 = [25,12,30,18,6,23]
y2 = [16,28,23,11,9,14]
```

```
plt.plot(x,y1)
plt.show()
```

```
# Cada linha recebe uma cor automaticamente caso não especificado
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

---
### [**Configurando uma linha**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)

```
# Simplificada(Format String): só configura Tipo de Marcador, Tipo de Linha e Cor da Linha (nessa ordem)
plt.plot(x, y1, '--r')

# Especificando cada parâmetro
plt.plot(x, y2, color='green', marker='*', linestyle='dotted', linewidth=3, markersize=14)

plt.show()
```

---
### **Título e rótulo dos eixos**


```
plt.plot(x,y1,'--r')
plt.plot(x,y2,'*:g')

plt.title('Gráfico')
plt.ylabel('Números')
plt.xlabel('Letras')

plt.show()
```

---
### [**Legenda**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html)

```
# Parâmetro "label" define a legenda de cada linha
plt.plot(x, y1, '--r', label='Linha 1')
plt.plot(x, y2, '*:g', label='Linha 2')

# Função "legend" cria a legenda
plt.legend()

plt.title('Gráfico')
plt.ylabel('Números')
plt.xlabel('Letras')

plt.show()
```

```
# Outra forma
plt.plot(x, y1, '--r')
plt.plot(x, y2, '*:g')

plt.legend(['Linha 1','Linha 2'])

plt.title('Gráfico')
plt.ylabel('Números')
plt.xlabel('Letras')

plt.show()
```

---
### **Texto**

```
plt.plot(x, y1, '--r', label='Linha 1')
plt.plot(x, y2, '*:g', label='Linha 2')

plt.title('Gráfico')
plt.ylabel('Números')
plt.xlabel('Letras')
plt.legend()

# X, Y, Texto
plt.text(0, 10, 'Dentro')
plt.text(6, 15, 'Fora') 
plt.text(0.7, 29, 'Ponto 2')
plt.text(2.7, 9, 'Ponto 4')

plt.show()
```

---
### **Tamanho da fonte**

```
plt.plot(x, y1, '--r', label='Linha 1')
plt.plot(x, y2, '*:g', label='Linha 2')

plt.title('Gráfico', fontsize=30)
plt.ylabel('Números', fontsize=18)
plt.xlabel('Letras', fontsize=12)
plt.legend(fontsize=16)
plt.text(0.7, 29, 'Ponto 2', fontsize=6)
plt.text(2.7, 9, 'Ponto 4', fontsize=18)

plt.show()
```

### [**Outros tipos de gráficos**](https://matplotlib.org/tutorials/introductory/sample_plots.html)

```
plt.scatter(x, y1, s=[100,25,50,500,250,1000], color='red')
plt.show()
```
